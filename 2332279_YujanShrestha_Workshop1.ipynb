```python
# Task 1: Load and display a grayscale image
image_path = "/content/drive/MyDrive/Ai&Ml/cameraman.png"
gray_image = Image.open(image_path).convert("L")  # Convert to grayscale
plt.figure(figsize=(5,5))
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

2. Extract and display the middle section of the image (150 pixels).
 Extract a 150 pixel section from the center of the image using NumPy array slicing.
 Display this cropped image using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

gray_array = np.array(gray_image)

center_y, center_x = gray_array.shape[0] // 2, gray_array.shape[1] // 2

crop_size = 150
half_crop = crop_size // 2


cropped_section = gray_array[center_y - half_crop:center_y + half_crop, center_x - half_crop:center_x + half_crop]

plt.figure(figsize=(5, 5))
plt.imshow(cropped_section, cmap="gray")
plt.title("Middle Section (150x150 pixels)")
plt.axis("off")
plt.show()
```

3. Apply a simple threshold to the image (e.g., set all pixel values below 100 to 0).
 Apply a threshold to the grayscale image: set all pixel values below 100 to 0, and all values
above 100 to 255 (creating a binary image).
 Display the resulting binary image.

```python
# Task 3: Apply a simple threshold to the image
binary_image = np.where(gray_array < 100, 0, 255).astype(np.uint8)  # Apply threshold
plt.figure(figsize=(5,5))
plt.imshow(binary_image, cmap="gray")
plt.title("Binary Image (Thresholded)")
plt.axis("off")
plt.show()
```

# **Exercise Three**

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

1. Load and Prepare Data:
 Fetch an image of you choice.{If colour convert to grayscale}
 Center the dataset - Standaridze the Data.
 Calculate the covaraince matrix of the Standaridze data.

```python
# Task 1: Load and display a grayscale image
image_path = "/content/drive/MyDrive/Ai&Ml/lena_gray.gif"
gray_image = Image.open(image_path).convert("L")  # Convert to grayscale
plt.figure(figsize=(5,5))
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

2. Eigen Decomposition and Identifying Principal Components:
 Compute Eigen Values and Eigen Vectors.
 Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
 Identify the Principal Components with the help of cumulative Sum plot.

```python
# Convert grayscale image to NumPy array
gray_array = np.array(gray_image)
```

```python
# Standardize the data
mean = np.mean(gray_array, axis=0)
std = np.std(gray_array, axis=0)
standardized_data = (gray_array - mean) / std
```

```python
# Compute the covariance matrix
cov_matrix = np.cov(standardized_data, rowvar=False)
```

```python
# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
```

```python
# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# Identify principal components with cumulative sum plot
explained_variance_ratio = np.cumsum(eigenvalues) / np.sum(eigenvalues)
plt.figure(figsize=(6,4))
plt.plot(explained_variance_ratio, marker='o', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance Plot")
plt.grid()
plt.show()
```

```python
# Function to reconstruct image using k principal components
def reconstruct_image(k):
    top_k_eigenvectors = eigenvectors[:, :k]
    transformed_data = np.dot(standardized_data, top_k_eigenvectors)
    reconstructed_data = np.dot(transformed_data, top_k_eigenvectors.T)
    reconstructed_image = (reconstructed_data * std) + mean  # Reverse standardization
    return np.clip(reconstructed_image, 0, 255).astype(np.uint8)
```

```python
# Experiment with different k values
k_values = [10, 50, 100, 200]
fig, axes = plt.subplots(1, len(k_values), figsize=(15,5))
for i, k in enumerate(k_values):
    reconstructed_image = reconstruct_image(k)
    axes[i].imshow(reconstructed_image, cmap="gray")
    axes[i].set_title(f"k={k}")
    axes[i].axis("off")
plt.show()
```
